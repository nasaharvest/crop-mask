```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('local_model_tests.csv')
```

```python
df.tail()
```

```python
df['Batch_size']=df['Batch_size'].astype('str')
```

```python
Namibia_df=df.query("Model_name=='Namibia_North_2020'")
Rwanda_df=df.query("Model_name=='Rwanda_2019'")
```

```python
#Plotting the CPU and memory percentage for Namibia_North_2020 model
figure,axs = plt.subplots(1,2, figsize=(15, 5), facecolor='w', edgecolor='k')
figure.suptitle('Namibia_North_2020 Peak CPU and memory vs batch_size')

axs = axs.flatten()
axs[0].plot(Namibia_df['Batch_size'],Namibia_df['Peak_cpu_percent'])
axs[0].set_ylabel('Peak CPU percentage')
axs[0].legend(['Nambia_North_2020'])
axs[0].legend(['Peak_system_CPU_percentage'])

axs[1].plot(Namibia_df['Batch_size'],Namibia_df['Peak_memory_percent'],color="red")
axs[1].legend(['Peak_system_memory_percentage'])
axs[1].set_ylabel('Peak Memory percentage')
```

```python
#Plotting the CPU and memory percentage for Rwanda_2019 model
figure,axs = plt.subplots(1,2, figsize=(15, 5), facecolor='w', edgecolor='k')
figure.suptitle('Rwanda_2019 Peak CPU and memory vs batch size')

axs = axs.flatten()
axs[0].plot(Rwanda_df['Batch_size'],Rwanda_df['Peak_cpu_percent'])
axs[0].set_ylabel('Peak CPU percentage')
axs[0].legend(['Peak_system_CPU_percentage'])

axs[1].plot(Rwanda_df['Batch_size'],Rwanda_df['Peak_memory_percent'],color="red")
axs[1].set_ylabel('Peak Memory percentage')
axs[1].legend(['Peak_system_memory_percentage'])
```

```python
#Plotting the boxplot of eeg sensor related columns
figure,axs = plt.subplots(1,2, figsize=(15, 5), facecolor='w', edgecolor='k')
figure.suptitle('Peak process CPU vs Batch size')

axs = axs.flatten()
axs[0].plot(Namibia_df['Batch_size'],Namibia_df['peak_proc_cpu'])
axs[0].legend(['Namibia_North_2020'])

axs[1].plot(Rwanda_df['Batch_size'],Rwanda_df['peak_proc_cpu'],color="red")
axs[1].legend(['Rwanda_2019'])
```

```python
#Plotting the boxplot of eeg sensor related columns
figure,axs = plt.subplots(1,2, figsize=(15, 5), facecolor='w', edgecolor='k')
figure.suptitle('Peak process memory vs Batch size')

axs = axs.flatten()
axs[0].plot(Namibia_df['Batch_size'],Namibia_df['peak_proc_mem'])
axs[0].legend(['Namibia_North_2020'])

axs[1].plot(Rwanda_df['Batch_size'],Rwanda_df['peak_proc_mem'],color="red")
axs[1].legend(['Rwanda_2019'])
```
