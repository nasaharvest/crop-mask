```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Malawi_2020_September.csv')
```

```python
df.head()
```

```python
df['No_of_requests']=df['No_of_requests'].astype('str')
```

```python
#Plotting the boxplot of eeg sensor related columns
figure,axs = plt.subplots(1,2, figsize=(10, 3), facecolor='w', edgecolor='k')
figure.suptitle('Malawi_Sept_2020 peak cpu and memory percentage variation according to the batch size')

axs = axs.flatten()
axs[0].plot(df['No_of_requests'],df['Peak_cpu_percent'])
axs[0].legend(['Peak_cpu_percent'])
axs[1].plot(df['No_of_requests'],df['Peak_memory_percent'],'y')
axs[1].legend(['Peak_memory_percent'])
```
